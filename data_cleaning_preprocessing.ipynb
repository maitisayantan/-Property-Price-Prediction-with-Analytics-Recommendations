```python
import re

def convert_to_sqft(x):
    
    s = str(x).strip()

    # Case 1: Range
    if '-' in s:
        try:
            low, high = map(float, s.split('-'))
            return (low + high) / 2
        except:
            return np.nan

    # Case 2: Extract number and optional unit
    match = re.match(r'([\d\.]+)\s*(.*)', s)
    if match:
        value = float(match.group(1))
        unit = match.group(2).strip().lower()

        if unit in ['sq. meter', 'sq. meters', 'sqm']:
            return value * 10.7639
        elif unit in ['sq. yard', 'sq. yards', 'sqyards', 'sqyards']:
            return value * 9.0
        elif unit in ['acre', 'acres']:
            return value * 43560.0
        elif unit in ['guntha', 'gunthas']:
            return value * 1089.0
        elif unit in ['cent', 'cents']:
            return value * 435.6
        else:
            return value  # assume already in sqft
    
    return np.nan

# Apply conversion
df['total_sqft'] = df['total_sqft'].apply(convert_to_sqft).round(1)
```
